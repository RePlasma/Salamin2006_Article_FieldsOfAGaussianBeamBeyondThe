# Figure 2 a

```python
eps = 0.8
lbd = 1
W0 = lbd/(pi*eps) #[\mu m] 
print("eps =",eps)
xlst = np.linspace(-5,+5,200) * W0
Exlst0 = np.array([ ExFocalPlane0(x, 0, 0, lbd, W0, 1) for x in xlst])
Exlst4 = np.array([ ExFocalPlane4(x, 0, 0, lbd, W0, 1) for x in xlst])
Exlst10 = np.array([ ExFocalPlane10(x, 0, 0, lbd, W0, 1) for x in xlst])
Exlst11_full = np.array([ ExyzBxyz(x, 0, 0, lbd, W0, 1)[0] for x in xlst])

plt.plot(xlst/W0, Exlst0, label=r'$\varepsilon^0$')
plt.plot(xlst/W0, Exlst4, label=r'$\varepsilon^{4}$')
plt.plot(xlst/W0, Exlst10, label=r'$\varepsilon^{10}$')
plt.plot(xlst/W0, Exlst11_full, '--', label=r'$\varepsilon^{11}$ Appendix A')
plt.xlim([-5,5])
plt.ylim([-0.2,1.4])
plt.legend()
plt.xlabel(r'$x/W_0$')
plt.ylabel(r'$E_x/E_0$')
plt.show()
```

# Figure 2 b

```python
# notice the very small value of peak Ez component
eps = 0.8
lbd = 1
W0 = lbd/(pi*eps) #[\mu m] 
print("eps =",eps)
xlst = np.linspace(-5,+5,200) * W0
Ezlst1 = np.array([ EzFocalPlane1(x, 0, 0, lbd, W0, 1) for x in xlst])
Ezlst5 = np.array([ EzFocalPlane5(x, 0, 0, lbd, W0, 1) for x in xlst])
Ezlst11 = np.array([ EzFocalPlane11(x, 0, 0, lbd, W0, 1) for x in xlst])
Ezlst11_full = np.array([ ExyzBxyz(x, 0, 0, lbd, W0, 1)[2] for x in xlst])

plt.plot(xlst/W0, Ezlst1, label=r'$\varepsilon$')
plt.plot(xlst/W0, Ezlst5, label=r'$\varepsilon^{5}$')
plt.plot(xlst/W0, Ezlst11, label=r'$\varepsilon^{11}$')
plt.plot(xlst/W0, Ezlst11_full, '--', label=r'$\varepsilon^{11}$ Appendix A')
plt.xlim([-5,5])
#plt.ylim([np.min(Ezlst11), np.max(Ezlst11)])
plt.legend()
plt.xlabel(r'$x/W_0$')
plt.ylabel(r'$E_z/E_0$')
plt.show()
```
